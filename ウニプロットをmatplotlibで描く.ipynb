```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
import math
x = np.linspace(0, 2.0 * math.pi, 20)
y = np.sin(x)
```

```python
from collections import OrderedDict
markers = OrderedDict([
    ('*', [200, 0, 0]),
    ('v', [100, 0, -0.02]),
    ('<', [70, -0.05, -0.01]),
    ('>', [70, 0.05, -0.01]),
    (4, [70, -0.035, 0.02]),
    (5, [70, 0.035, 0.02]),
])
for m, s in markers.items():
    plt.scatter(x + s[1], y + s[2], marker=m, c="black", s=s[0])
    plt.show()
```
