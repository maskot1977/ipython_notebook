```python
from rdkit import Chem
from rdkit.Chem import AllChem
from IPython.display import Image
from rdkit.Chem import Draw
```

```python
rxn1 = AllChem.ReactionFromSmarts('[C:1](=[O:2])-[OD1].[N!H0:3]>>[C:1](=[O:2])[N:3]')
substrates1 = ['OC(=O)CC(N)C(=O)O', 'OC(=O)c1ccc(N)c(C(=O)O)c1']
substrates2 = ['NCC(=O)N', 'NCN']
```

```python
ms1 = [Chem.MolFromSmiles(substrate) for substrate in substrates1]
#for m in ms: tmp=AllChem.Compute2DCoords(m)
img=Draw.MolsToGridImage(ms1,molsPerRow=4,subImgSize=(200,200), legends=substrates1)
img.save('grid1.png')
Image(filename="grid1.png")
```

```python
ms2 = [Chem.MolFromSmiles(substrate) for substrate in substrates2]
#for m in ms: tmp=AllChem.Compute2DCoords(m)
img=Draw.MolsToGridImage(ms2,molsPerRow=4,subImgSize=(200,200), legends=substrates2)
img.save('grid2.png')
Image(filename="grid2.png")
```

```python
all_products = []
for m1 in ms1:
    for m2 in ms2:
        all_products.append(rxn1.RunReactants((m1,m2)))
```

```python
smiles_products = []
for all_ps in all_products:
    for all_p in all_ps:
        for m in all_p:
            Chem.SanitizeMol(m)
            smile = Chem.MolToSmiles(m)
            if smile not in smiles_products:
                smiles_products.append(smile)
```

```python
ms_products = [Chem.MolFromSmiles(product) for product in smiles_products]
img=Draw.MolsToGridImage(ms_products, molsPerRow=4, subImgSize=(200,200), legends=smiles_products)
img.save('grid.png')
Image(filename="grid.png")
```
